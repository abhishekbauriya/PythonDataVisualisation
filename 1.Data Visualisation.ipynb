```python
# Line Plot
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
# Normal Plot

plt.plot(x,y1,color='red',label="Apple")
plt.plot(x,y2,color='green',label="Kiwi")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend()
plt.show()
```

```python
# Dotted Plot with different styles(*,^,o)

plt.plot(x,y1,color='red',label="Apple",marker='o')
plt.plot(x,y2,color='green',label="Kiwi",linestyle="dashed",marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fuits ovet Time")
plt.legend()
plt.show()
```

# Data Visualisation Challenge

Given a Movies Dataset, your tasks is to make a visualisation that -

- Plots the length of MOvie Title Names on X-axis
- Frequency COunts on Y-axis i.e. Number of MOvies having 'x' characters in their title

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read a CSV File
df = pd.read_csv("movie_metadata.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
titles = df.get('movie_title')
```

```python
print(titles)
```

```python
#TO get into list format

titles = list(df.get('movie_title'))
```

```python
print(type(titles))
```

```python
print(titles[:5])
```

```python
print(titles[0][:-1])
```

```python
#Making dictionary 

freq_titles = {}

for title in titles:
    length = len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
#print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Freq Count")
plt.title("Movie Data Visualisation Problem")
plt.show()
```
